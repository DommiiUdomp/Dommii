```python
import pandas as pd
```

```python
#download from zip
import zipfile
import io
import requests
```

```python
# file URL
url = "https://databank.worldbank.org/data/download/hnp_stats_csv.zip"

# download zip
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))

# get file names
csvfiles = []
for file in z.namelist():
  #df = pd.read_csv(file)
  csvfiles.append(file)
print(csvfiles)

df_all = []
for csvfile in csvfiles:
  df = pd.read_csv(z.open(csvfile))
  df_all.append(df) 

df = pd.concat(df_all,ignore_index=True)
df.head()
```
